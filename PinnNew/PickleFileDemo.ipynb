The purpose of this file is to explain how to use previously trained PINNs in form of pickles files or save them in one.

The easiest way is go to the desired file (e.g. KdVOneSoliton.ipynb), and run everything up until initialization and compilation of the PINN (included). Then instead of proceeding to train run the following sell to set the weights and the history file (if applicable).

```python
with open('PickleFiles/KdVB/Alpha0-5/KdVBalpha1-5Time10-3x32SmallAmpl.pickle', 'rb') as f:
    my_weights = pickle.load(f)
pinn.backbone.set_weights(my_weights)
with open('PickleFiles/KdVB/Alpha0-5/HistoryKdVBalpha0-5Time10-3x32SmallAmpl.pickle', 'rb') as f:
  history = pickle.load(f)
```

Things to ensure: <br>


*  pickle is imported
*   the path to the pickle file is correct
* the network architecture of the pickle file matches that of the file loaded
* ask the pinn for predictions of the correct domain (that of the pickle file)



In order to save the weight and history in form of pickle files simply run the following with desired path to the new files

```python
with open('PickleFiles/KdV/OneSoliton/HistoryOneSoliton.pickle', 'wb') as fp:
    pickle.dump(history, fp)

current_weights = pinn.backbone.get_weights()
with open('PickleFiles/KdV/OneSoliton/KdVOneSolitonTime10-3x32.pickle', 'wb') as f:
    pickle.dump(current_weights, f)
```
